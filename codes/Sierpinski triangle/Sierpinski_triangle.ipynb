```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import cm
```

```python
# creating the triangle 
v1 = np.array([0,0])
v2 = np.array([1,0]) 
v3 = np.array([1/2,np.sqrt(3)/4])   
```

```python
#Generating the initial random point inside the traingle

def random_point_in_triangle(v1, v2, v3):
    # Generate 3 random numbers between 0 and 1
    r1, r2, r3 = np.random.random(), np.random.random(), np.random.random()
    # Normalize them
    s1 = r1/(r1+ r2 + r3)
    s2 = r2/(r1+ r2 + r3)
    s3 = r3/(r1+ r2 + r3)
    # Using random convex combination 
    x = s1 * v1[0] + s2 * v2[0] + s3 * v3[0]
    y = s1 * v1[1] + s2 * v2[1] + s3 * v3[1]
    
    return np.array([x, y])
```

```python
v_x = np.array([v1[0],v2[0],v3[0],v1[0]])
v_y = np.array([v1[1],v2[1],v3[1],v1[1]])
```

```python
initial_piont = random_point_in_triangle(v1, v2, v3)
plt.plot(v_x, v_y, color = 'b')
plt.scatter(initial_piont[0],initial_piont[1], color = 'r', s=5)
plt.xlim(min(v_x)-0.05, max(v_x)+0.05)
plt.ylim(min(v_y)-0.05, max(v_y)+0.05)
plt.tick_params(axis='both', which='both', length=0)
plt.xticks([])
plt.yticks([]);
```

```python
vertices_x= np.array([v1[0],v2[0],v3[0]])
vertices_y = np.array([v1[1],v2[1],v3[1]]) 
```

```python
def serpinskis_points(N, initial_point):
    points_x = np.array([])
    points_y = np.array([])
    vertices_chosen = np.array([])
    
    first_vetex_num =  np.random.randint(3)
    
    vertex = np.array([vertices_x[first_vetex_num], vertices_y[first_vetex_num]])
    vertices_chosen = np.append(vertices_chosen,first_vetex_num)
    first_point = 0.5* (initial_point+ vertex) 
    points_x = np.append(points_x,first_point[0])
    points_y = np.append(points_y,first_point[1])
    for i in range(N):
        new_init_point_x = points_x[i] #choosing the x coordinate of the last point
        new_init_point_y = points_y[i] #choosing the y coordinate of the last point
        new_vertex_number = np.random.randint(3) # Choosing a new random vertex number
        vertices_chosen = np.append(vertices_chosen,new_vertex_number) # Saving the new random vertex number
        new_point_x = 0.5*(new_init_point_x + vertices_x[new_vertex_number]) # creating a new point x coordinate
        new_point_y = 0.5*(new_init_point_y + vertices_y[new_vertex_number]) # creating a new point y coordinate
        
        points_x = np.append(points_x,new_point_x)
        points_y = np.append(points_y,new_point_y)
    return points_x, points_y, vertices_chosen
        
```

```python
fontsize= 16
ticksize = 14
figsize = (7, 7)
params = {"text.usetex": True,
    'font.family':'serif',
    "figure.figsize":figsize, 
    'figure.dpi': 80,
    'figure.edgecolor': 'k',
    'font.size': fontsize, 
    'axes.labelsize': fontsize,
    'axes.titlesize': fontsize,
    'xtick.labelsize': ticksize,
    'ytick.labelsize': ticksize
}
plt.rcParams.update(params) 
```

```python
x_points , y_points, vertices = serpinskis_points(10000, initial_piont)
```

```python
plt.plot(v_x, v_y, color = 'b')
plt.scatter(x_points,y_points, color = 'b', s=0.1)
plt.xlim(min(v_x)-0.05, max(v_x)+0.05)
plt.ylim(min(v_y)-0.05, max(v_y)+0.05)
plt.tick_params(axis='both', which='both', length=0)
plt.xticks([])
plt.yticks([]);
```
